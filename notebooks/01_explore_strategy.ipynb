# 01 - Explore strategy

Quick template for poking at a strategy on the bundled fixture.
Paper-trading-only project. See `REALITY-CHECK.md`.

```python
from pathlib import Path

from daytrade_ai.backtest.engine import BacktestEngine
from daytrade_ai.data.csv_source import CSVDataSource
from daytrade_ai.reporting.report import format_text_report
from daytrade_ai.strategies import get_strategy, list_strategies

df = CSVDataSource(Path("../tests/fixtures/btc_sample.csv")).fetch("BTC/USDT", "1h")
print(list_strategies())
df.head()
```

```python
strat = get_strategy("sma_cross", fast=5, slow=20)
engine = BacktestEngine(initial_cash=10_000, fee_bps=10, slippage_bps=5)
result = engine.run(df, strat)
print(format_text_report(result, title="sma_cross :: BTC/USDT :: 1h"))
```
